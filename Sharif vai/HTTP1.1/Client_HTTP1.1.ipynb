```python
import time
import requests
import numpy as np

def transfer_http1_1(file_size, num_runs, source_url, destination_url):
    total_application_data_transferred = 0
    for _ in range(num_runs):
        response = requests.get(source_url)
        total_application_data_transferred += len(response.content) + len(response.headers)  # Include both content and headers
        with open(destination_url, 'wb') as f:
            f.write(response.content)
    return total_application_data_transferred / (num_runs * file_size)

source_url_http1_1 = 'http://192.168.1.11:8000/B_10kB.txt'  # Adjust server IP and port if needed
destination_url_http1_1 = 'B_10kB.txt'  # Save in local directory

num_runs = 1000

total_ratio_http1_1 = transfer_http1_1(1000 * 1024, num_runs, source_url_http1_1, destination_url_http1_1)
#total_ratio_http1_1 *= 3  # Multiply by 3 for BitTorrent

print("Total Application Data Transferred Ratio (HTTP/1.1):", total_ratio_http1_1)
```
